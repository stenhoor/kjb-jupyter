```python
import pandas as pd
```

```python
xl = pd.read_excel(
    "Sample Data.xlsx", 
    sheet_name=pd.ExcelFile("Sample Data.xlsx").sheet_names, 
    index_col=0
)
```

```python
for sheet in xl.keys():
    print('Key {}'.format(sheet))
    xl[sheet] = xl[sheet].T.assign(sheet_name=sheet).set_index(['sheet_name'], append=True)
```

```python
df = pd.concat(xl.values())
```
